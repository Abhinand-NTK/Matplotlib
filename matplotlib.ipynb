```python
import numpy as np
```

```python
x = np.arange(0,10,2)
y = np.arange(11,21,2)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
plt.savefig('Test_scatter.png')
```
